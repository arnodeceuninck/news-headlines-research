# Extra features
Up to now, we've only mentioned numbers generated by a vectorizer as extra features. It's also possible to add even more features to it. I often base my extra features on other natural language processing libraries.
## SpaCy
SpaCy is a Python library for natural language processing. It can be used to extract features from text, such as part-of-speech tags, named entities, and dependency parse trees. I will use it later. More info about it can be found [here](https://spacy.io/).
